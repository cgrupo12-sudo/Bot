# Task
Genera una gráfica interactiva de la inflación anual de todos los países a partir del archivo "Inflacion.xlsx", permitiendo al usuario seleccionar los países y el rango de fechas a visualizar.

## Cargar los datos

### Subtask:
Cargar los datos del archivo Excel `Inflacion.xlsx` en un DataFrame de pandas.


**Reasoning**:
Import the pandas library and load the Excel file into a DataFrame, then display the first few rows.



```python
import pandas as pd

df_inflacion = pd.read_excel("Inflacion.xlsx")
display(df_inflacion.head())
```

## Preprocesar los datos

### Subtask:
Limpiar y transformar los datos para que sean adecuados para la visualización. Esto podría incluir la conversión de fechas y el manejo de valores faltantes.


**Reasoning**:
Clean and transform the data as instructed, which involves ensuring the 'Año' column is an integer type, checking for and handling missing values in 'Inflación Anual', and confirming 'Inflación Anual' is numerical.



```python
# Ensure 'Año' is integer type
df_inflacion['Año'] = df_inflacion['Año'].astype(int)

# Check for missing values in 'Inflación Anual'
missing_inflation_values = df_inflacion['Inflación Anual'].isnull().sum()
print(f"Number of missing values in 'Inflación Anual': {missing_inflation_values}")

# Handle missing values - for this task, dropping rows with missing inflation is a reasonable approach
if missing_inflation_values > 0:
    df_inflacion.dropna(subset=['Inflación Anual'], inplace=True)
    print("Dropped rows with missing 'Inflación Anual' values.")

# Ensure 'Inflación Anual' is numerical - read_excel usually handles this, but a check is good practice
if not pd.api.types.is_numeric_dtype(df_inflacion['Inflación Anual']):
    # Attempt to convert if not already numerical
    df_inflacion['Inflación Anual'] = pd.to_numeric(df_inflacion['Inflación Anual'], errors='coerce')
    # After coercion, check and handle any new NaNs created by the conversion
    missing_inflation_values_after_coerce = df_inflacion['Inflación Anual'].isnull().sum()
    if missing_inflation_values_after_coerce > 0:
         df_inflacion.dropna(subset=['Inflación Anual'], inplace=True)
         print(f"Dropped {missing_inflation_values_after_coerce} rows with non-numeric 'Inflación Anual' values after coercion.")

display(df_inflacion.head())
```

## Crear la gráfica inicial

### Subtask:
Generar una gráfica inicial de la inflación anual para todos los países utilizando Matplotlib.


**Reasoning**:
Generate an initial plot of annual inflation for all countries using Matplotlib as requested by the instructions.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
ax = plt.axes()

for country in df_inflacion['País'].unique():
    country_data = df_inflacion[df_inflacion['País'] == country]
    ax.plot(country_data['Año'], country_data['Inflación Anual'], label=country)

plt.title('Inflación Anual por País')
plt.xlabel('Año')
plt.ylabel('Inflación Anual')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

## Implementar la selección de países y fechas

### Subtask:
Añadir elementos interactivos para permitir al usuario seleccionar los países y el rango de fechas a visualizar.


**Reasoning**:
Import necessary libraries for interactive widgets and display, create the interactive widgets for country selection and year range, define the update function, and link them using ipywidgets.interactive, finally display the widgets and the initial plot.



```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

# Get unique countries and year range
all_countries = df_inflacion['País'].unique().tolist()
min_year = df_inflacion['Año'].min()
max_year = df_inflacion['Año'].max()

# Create widgets
country_selector = widgets.SelectMultiple(
    options=all_countries,
    value=all_countries,
    description='Select Countries:',
    disabled=False
)

year_slider = widgets.IntRangeSlider(
    value=[min_year, max_year],
    min=min_year,
    max=max_year,
    step=1,
    description='Select Year Range:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# Define the update function
def update_plot(selected_countries, year_range):
    plt.figure(figsize=(12, 8))
    ax = plt.axes()

    start_year, end_year = year_range

    filtered_df = df_inflacion[(df_inflacion['País'].isin(selected_countries)) &
                               (df_inflacion['Año'] >= start_year) &
                               (df_inflacion['Año'] <= end_year)]

    for country in selected_countries:
        country_data = filtered_df[filtered_df['País'] == country]
        ax.plot(country_data['Año'], country_data['Inflación Anual'], label=country)

    plt.title('Inflación Anual por País')
    plt.xlabel('Año')
    plt.ylabel('Inflación Anual')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()

# Link widgets and update function
interactive_plot = widgets.interactive(update_plot, selected_countries=country_selector, year_range=year_slider)

# Display widgets and the plot
display(interactive_plot)
```
